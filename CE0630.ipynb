```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./income.csv')
df.head()
```

```python
df.dtypes
```

```python
df.keys()
```

```python
df.describe()
```

```python
for col in df.columns:
    if df[col].dtype == 'int64':
        print('INT64',col, ':', df[col].value_counts().index.tolist())
    if df[col].dtype == 'object':
        print('OBJECT',col, ':', df[col].value_counts().index.tolist())
```

```python
df.isna()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
n_dup = df.duplicated().sum()

n_na = df.isna().sum()

print('No. of duplicates : ',n_dup,'\nNo. of null values :',n_na)
```

```python
import matplotlib.pyplot as plt
plt.hist(df['SalStat'])
plt.show()

import seaborn as sns
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.show()

contingency_table = pd.crosstab(df['EdType'], df['SalStat'])
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(contingency_table)
print('p-value:', p)

```
